```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create a 3D NumPy array of shape (3x3x3) to simulate an RGB image
# Values range from 0 to 255 (like pixel intensities)
image = np.random.randint(0, 256, (3, 3, 3), dtype=np.uint8)
print("Original 3D RGB Image Array:\n", image)

# 2. Extract Red, Green, Blue layers (channel-wise slicing)
red_channel   = image[:, :, 0]
green_channel = image[:, :, 1]
blue_channel  = image[:, :, 2]

print("\nRed Channel:\n", red_channel)
print("\nGreen Channel:\n", green_channel)
print("\nBlue Channel:\n", blue_channel)

# 3. Compute average pixel intensity per channel
red_avg   = np.mean(red_channel)
green_avg = np.mean(green_channel)
blue_avg  = np.mean(blue_channel)

print("\nAverage Intensity per Channel:")
print(f"Red   = {red_avg:.2f}")
print(f"Green = {green_avg:.2f}")
print(f"Blue  = {blue_avg:.2f}")

# 4. Add a scalar to brighten the image (clip values at 255)
brightened_image = np.clip(image + 50, 0, 255)  # add 50 to each pixel
print("\nBrightened Image Array:\n", brightened_image)

# 5. Visualization using matplotlib
fig, axs = plt.subplots(1, 4, figsize=(12, 4))

# Original
axs[0].imshow(image)
axs[0].set_title("Original Image")
axs[0].axis("off")

# Red channel visualization
axs[1].imshow(red_channel, cmap="Reds")
axs[1].set_title("Red Channel")
axs[1].axis("off")

# Green channel visualization
axs[2].imshow(green_channel, cmap="Greens")
axs[2].set_title("Green Channel")
axs[2].axis("off")

# Blue channel visualization
axs[3].imshow(blue_channel, cmap="Blues")
axs[3].set_title("Blue Channel")
axs[3].axis("off")

plt.show()

# 6. Brightness change discussion
print("\nDiscussion:")
print("We increased brightness by adding 50 to every pixel value.")
print("Since RGB values must stay in [0, 255], we clipped them.")
print("This makes the image look brighter while preserving valid color values.")
```

```python
import numpy as np
# 1. Create a 2D array and add a 1D array
A = np.array([[1, 2, 3],
              [4, 5, 6]])   # Shape (2,3)
b = np.array([10, 20, 30])  # Shape (3,)
print("2D Array A:\n", A)
print("Shape of A:", A.shape)
print("\n1D Array b:", b)
print("Shape of b:", b.shape)
# Broadcasting: (2,3) + (3,) -> (2,3)
C = A + b
print("\nResult of A + b:\n", C)
# 2. Broadcasting explanation
print("\nBroadcasting Explanation (2D + 1D):")
print("A has shape (2,3)")
print("b has shape (3,) -> treated as (1,3)")
print("b is broadcast along rows → result shape (2,3)")
# 3. 3D array with scalar addition
image = np.ones((3,3,3), dtype=int) * 50   # Shape (3,3,3)
scalar = 100
print("\n3D Array (image):\n", image)
print("Shape of image:", image.shape)
brightened = image + scalar
print("\nResult of image + scalar (100):\n", brightened)
print("\nBroadcasting Explanation (3D + scalar):")
print("Image shape: (3,3,3)")
print("Scalar shape: () -> broadcast to (3,3,3)")
# 4. Using np.newaxis / reshape for proper broadcasting
x = np.array([1, 2, 3])      # Shape (3,)
y = np.array([10, 20])       # Shape (2,)
print("\nOriginal arrays:")
print("x:", x, "Shape:", x.shape)
print("y:", y, "Shape:", y.shape)
# Add using newaxis
x_reshaped = x[np.newaxis, :]   # Shape (1,3)
y_reshaped = y[:, np.newaxis]   # Shape (2,1)
print("\nReshaped arrays:")
print("x reshaped:", x_reshaped, "Shape:", x_reshaped.shape)
print("y reshaped:\n", y_reshaped, "Shape:", y_reshaped.shape)
result = x_reshaped + y_reshaped
print("\nResult of broadcasting (x + y):\n", result)
print("Result shape:", result.shape)
print("\nBroadcasting Explanation (reshaped):")
print("x reshaped -> (1,3)")
print("y reshaped -> (2,1)")
print("Result shape: (2,3) after broadcasting")
# 5. Invalid broadcasting case
print("\nInvalid Broadcasting Example:")
p = np.array([[1,2,3],[4,5,6]])   # Shape (2,3)
q = np.array([1,2])               # Shape (2,)

print("p shape:", p.shape)
print("q shape:", q.shape)

try:
    invalid_result = p + q
except ValueError as e:
    print("Error:", e)
```
