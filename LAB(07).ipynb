```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
world_population = pd.read_csv("world_population.csv")

# Extract population columns across years
population_cols = [col for col in world_population.columns if "Population" in col and "World" not in col]

# Sum all countries to get world total
pop_data = world_population[population_cols].sum().reset_index()
pop_data.columns = ["Year", "Population"]  # Fixed: corrected column assignment and quotes

# Clean up year column
pop_data["Year"] = pop_data["Year"].str.replace(" Population", "").astype(int)  # Fixed: corrected quotes

# Plot world population growth
plt.figure(figsize=(10, 6))  # Fixed: corrected function name and quotes
plt.plot(pop_data["Year"], pop_data["Population"], marker="o", color="blue", label="World Population")  # Fixed: corrected quotes
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Growth Over Years")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
aapl_stock = pd.read_csv("aapl.csv")

# Convert 'Date' column to datetime
aapl_stock["Date"] = pd.to_datetime(aapl_stock["Date"])

# Plot stock closing prices
plt.figure(figsize=(12, 6))
plt.plot(aapl_stock["Date"], aapl_stock["Close"], color="green", label="Apple Close Price")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Apple Stock Price Trend")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
